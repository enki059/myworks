```python
# 1)
def count_digits(a, b):
  a1, b1 = a, b
  while a1 > 0 and b1 > 0:
    a1 //= 10
    b1 //= 10
  return a if a1 > 0 else b

print(count_digits(14023, 12128))
```

```python
# 2)
def gcd_and_lcm(a, b):
  gcd = 1
  for i in range(2, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
      gcd = i
  lcm = a * b // gcd
  return (gcd, lcm)

print(gcd_and_lcm(12, 56))
```

```python
# 3)
def digits(a, count = 0):
  if a > 0:
    return digits(a // 10, count + 1)
  return count

print(digits(1000))
```
