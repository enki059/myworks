```python
class Document:
  def __init__(self, name, author, content):
    self.name = name
    self.author = author
    self.content = content

  def update_content(self, newContent):
    self.content = newContent

  def display_info(self):
    print(f"name = {self.name}, author = {self.author}, content = {self.content}")

class Recipe(Document):
  def __init__(self, name, author, content, ingr, time):
    super().__init__(name, author, content)
    self.ingr = ingr
    self.time = time

  def update_time(self, newtime):
    self.time = newtime

  def display_info(self):
    print("Recipe:", self.content, "Ingredients:", self.ingr, "Time:", self.time, sep='\n')

rec1 = Recipe("a1", "aa", "omg", "apple", "1h")
rec1.display_info()
rec1.update_content("aaaaaaaaa")
rec1.update_time("2h")
rec1.display_info()
```

```python
class Worker:
    def __init__(self, name, title):
        self.name = name
        self.title = title

    def get_name(self):
        return self.name

    def get_title(self):
        return self.title

    def display_info(self):
        print(f"Имя: {self.name}, Должность: {self.title}")

class Project:
    def __init__(self):
        self.workers = []

    def add_worker(self, student):
        self.workers.append(student)

    def display_workers(self):
        print("Список сотрудников в проекте:")
        for worker in self.workers:
            worker.display_info()

w1 = Worker("a1", "a")
w2 = Worker("b1", "b")
w3 = Worker("c1", "c")

pr = Project()
pr.add_worker(w1)
pr.add_worker(w2)
pr.add_worker(w3)

pr.display_workers()
```

```python
import abc
class ElectronicDevice(abc.ABC):
  @abc.abstractmethod
  def turnOn(self): pass

class Smartphone(ElectronicDevice):
  def turnOn(self):
    print("Смартфон включен")

class Laptop(ElectronicDevice):
  def turnOn(self):
    print("Ноутбук включен")

s1 = Smartphone()
l1 = Laptop()

s1.turnOn()
l1.turnOn()
```
