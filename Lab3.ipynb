```python
def invert(dict1):
  res = {value : [] for key, value in dict1.items()}
  for key, value in dict1.items():
    res[value].append(key)
  return res


d = {"Einst端rzende Neubauten": "Blixa Bargeld", "Kraftwerk": "Ralf H端tter", "Franz Ferdinand": "Alex Kapranos",
     "Lacrimosa": "Tilo Wolff", "Ton Steine Scherben": "Rio Reiser", "Talking Heads": "David Byrne",
     "Nick Cave and the Bad Seeds": "Blixa Bargeld", "FFS": "Alex Kapranos"}
print(invert(d))
```

```python
def findKeys(item, dict1):
  result = []
  for key, value in dict1.items():
    if item == value:
      result.append(key)
  return result


print(findKeys("Ralf H端tter", d))
print(findKeys("Blixa Bargeld", d))
```

```python
def sortLen(dict1):
  res = {len(x) : [] for x in dict1.keys()}
  for x in dict1.keys():
    res[len(x)].append(x)
  return res


print(sortLen(d))
```

```python
def maxVal(dict1):
  maxLen = 0
  resultKey = ""
  for key, value in dict1.items():
    tmpLen = len(value)
    if tmpLen > maxLen:
      maxLen = tmpLen
      resultKey = key
  return resultKey


print(maxVal(d))
```

```python
def evenOdd(set1):
  res0, res1 = [], []
  for x in set1:
    if len(x) % 2 == 0:
      res0.append(x)
    else:
      res1.append(x)
  return res0, res1


lyrics = """Radioaktivit辰t
F端r dich und mich in All entsteht
Radioaktivit辰t
Strahlt Wellen zum Empfangsger辰t
Radioaktivit辰t
Wenns um unsere Zukunft geht
Radioaktivit辰t
F端r dich und mich in All entsteht
Radioactivity
Is in the air for you and me
Radioactivity
Discovered by Madame Curie
Radioactivity
Tune in to the melody
Radioactivity
Is in the air for you and me"""

set1 = set(lyrics.split(sep="\n"))
print(evenOdd(set1))
```

```python
def isSubset(set1, set2):
  return True if set1 == set2.intersection(set1) else False


set1 = set([1, 2, 4])
set2 = set([3, 7, 8, 9])
set3 = set([8, 3])
print(isSubset(set1, set2), isSubset(set3, set2))
```

```python
def symDiff(set1, set2):
  intersect = set1.intersection(set2)
  unionSet = set1.union(set2)
  return unionSet.difference(intersect)


set1 = set([1, 3, 7, 4])
set2 = set([3, 7, 8, 9])

print(symDiff(set1, set2), set1.symmetric_difference(set2))
```

```python
from random import randrange

def unionTuple(tuple1, tuple2):
  return tuple(set(tuple1 + tuple2))


tuple1 = tuple(randrange(1, 10) for i in range(8))
tuple2 = tuple(randrange(1, 10) for i in range(6))

print("tuple1:", tuple1)
print("tuple2:", tuple2)
print("unionTuple:", unionTuple(tuple1, tuple2))
```

image.png

```python
def tupleCount(elem, tup1):
  count = 0
  for x in tup1:
    if x == elem:
      count += 1
  return count


tuple1 = tuple(randrange(1, 10) for i in range(15))
print(tuple1)
print(tupleCount(1, tuple1))
```

image.png

```python
def splitTuple(tup1):
  part1 = tuple(x for x in range((len(tup1) + 1) // 2))
  part2 = tuple(x for x in range((len(tup1) + 1) // 2, len(tup1)))
  return part1, part2


tuple1 = tuple(randrange(1, 10) for i in range(7))
tuple2 = tuple(randrange(1, 10) for i in range(6))
print(splitTuple(tuple1))
print(splitTuple(tuple2))
```

```python
def frequencyTuple(tuple1):
  result = {elem: list(tuple1).count(elem) for elem in tuple1}
  return result


tuple1 = tuple(randrange(1, 4) for i in range(7))
print(tuple1)
print(frequencyTuple(tuple1))
```

```python
def filterTuple(a, tuple1):
  res = []
  for x in tuple1:
    if x > a:
      res.append(x)
  return tuple(res)


tuple1 = tuple(randrange(1, 10) for i in range(15))
print(tuple1)
print(filterTuple(6, tuple1))
```

```python
def sumTuple(tuple1):
  summa = 0
  for x in tuple1:
    summa += x
  return summa


tuple1 = tuple(randrange(1, 10) for i in range(15))
print(tuple1)
print(sumTuple(tuple1))
```

```python
def multTuple(tuple1):
  mult = 1
  for x in tuple1:
    mult *= x
  return mult


tuple1 = tuple(randrange(1, 10) for i in range(15))
print(tuple1)
print(multTuple(tuple1))
```
