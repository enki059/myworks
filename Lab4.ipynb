```python
text = """Октябрь уж наступил — уж роща отряхает
Последние листы с нагих своих ветвей;
Дохнул осенний хлад — дорога промерзает.
Журча еще бежит за мельницу ручей,
Но пруд уже застыл; сосед мой поспешает
В отъезжие поля с охотою своей,
И страждут озими от бешеной забавы,
И будит лай собак уснувшие дубравы."""

symbol = input("Введите символ: ")
data = text.split(sep="\n")
for line in data:
  print(f'Символ {symbol} встречается в строке "{line.strip()}" {line.count(symbol)} раз')
```

```python
text = """aaaaaa bbbbbb ccc dddd
  eee bfffffff aaaa rrr
  dd ttt www eeeee
  sssssss bbb"""

count = 0
words = text.split()
for word in words:
  if word.startswith("b"):
    count += 1

print(f"Количество слов, начинающихся с b: {count}")
```

```python
from random import random

n = 10
mult = 1
with open("task1.txt", "w") as file:
  for i in range(n):
    num = random() * 10
    mult *= num
    file.write(f"{num}\n")

print(mult)
```

```python
from random import randrange

# заполняем файл данными
n = 10
maxLuggage = 5
maxWeight = 20
with open("Bagazh.txt", "w") as file:
  for i in range(n):
    file.write(f"{randrange(1, maxLuggage)} {randrange(1, maxWeight)}\n")

average = 0
countPas = 0
with open("Bagazh.txt", "r") as file:
  for line in file:
    luggage, weight = map(int, line.split())
    print(luggage, weight)
    average += luggage
    if luggage > 2:
      countPas += 1

print(f"количество пассажиров, имеющих более 2 вещей: {countPas}")

average /= n
countPas2 = 0
with open("Bagazh.txt", "r") as file:
  for line in file:
    luggage, weight = map(int, line.split())
    #print(luggage, weight)
    if luggage > average:
      countPas2 += 1

print(f"average: {average}")
print(f"количество пассажиров, количество вещей которых превосходит среднее число вещей: {countPas2}")
```

```python
ruCount = 0
numCount = 0
with open("task3.txt", "r") as file:
  text = file.read()
  for char in text:
    if 48 <= ord(char) <= 57:
      numCount += 1
    if 1040 <= ord(char) <= 1103:
      ruCount += 1

if ruCount > numCount:
  print("русских букв больше")
elif ruCount == numCount:
  print("русских букв и цифр равное количество")
else:
  print("цифр больше")
```
